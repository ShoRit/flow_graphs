```python
from collections import Counter
import pickle
```

```python
# not doing risec bc I accidentally overwrote it
datasets = ["japflow", "chemu", "mscorpus"]
```

```python
all_relations = []

for dataset in datasets:
    with open(f"/projects/flow_graphs/data/{dataset}/amr.pkl", "rb") as f:
        amrs = pickle.load(f)
    for document in amrs:
        for sentence in document:
            amr_graph = sentence["graph"]
            if amr_graph is None:
                continue
            relations = [triple[1] for triple in amr_graph.triples]
            all_relations.extend(relations)
```

```python
relation_set = set(all_relations)
relation_counter = Counter(all_relations)
```

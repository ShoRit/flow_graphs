```python
import json

from amrlib import load_stog_model
```

```python
amr_model = load_stog_model(device="cuda:0", batch_size=4)
```

```python
datasets = ["wsj", "risec", "japflow", "chemu", "mscorpus"]
dataset = "risec"
```

```python
with open(f"/projects/flow_graphs/data/{dataset}/train.json") as f:
    instances = json.load(f)
```

```python
sents = [instance["text"] for instance in instances]
```

```python
result = amr_model.parse_sents(sents, disable_progress=False, return_penman=True)
```
